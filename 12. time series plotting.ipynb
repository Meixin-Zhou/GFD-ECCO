```python
import matplotlib.pyplot as plt
import matplotlib.ticker
import numpy as np
```

```python
#load file
deg10 = np.loadtxt("10.csv", delimiter =",")
```

```python
month = deg10[0]
amoc = deg10[1]

#plot the blue curve
plt.figure(figsize=(30,10))
plt.plot(month, amoc, color='blue', linewidth=4)
plt.xlim(0, 26*12)
plt.ylim(0, 28)

#plot horizontal dotted lines
a = [0, 26*12-1]
b = [10,10]
c = [20,20]
plt.plot(a,b,linestyle='dotted', color='black')
plt.plot(a,c,linestyle='dotted', color='black')

#plot vertical dotted lines
d = [0,22]
a1 = [24,24]
a2 = [48,48]
a3 = [3*24, 3*24]   
a4 = [4*24, 4*24] 
a5 = [5*24, 5*24] 
a6 = [6*24, 6*24] 
a7 = [7*24, 7*24] 
a8 = [8*24, 8*24] 
a9 = [9*24, 9*24] 
a10 = [10*24, 10*24] 
a11 = [11*24, 11*24] 
a12 = [12*24, 12*24] 
plt.plot(a1,d,linestyle='dotted', color='black')
plt.plot(a2,d,linestyle='dotted', color='black')
plt.plot(a3,d,linestyle='dotted', color='black')
plt.plot(a4,d,linestyle='dotted', color='black')
plt.plot(a5,d,linestyle='dotted', color='black')
plt.plot(a6,d,linestyle='dotted', color='black')
plt.plot(a7,d,linestyle='dotted', color='black')
plt.plot(a8,d,linestyle='dotted', color='black')
plt.plot(a9,d,linestyle='dotted', color='black')
plt.plot(a10,d,linestyle='dotted', color='black')
plt.plot(a11,d,linestyle='dotted', color='black')
plt.plot(a12,d,linestyle='dotted', color='black')

#set labels
yticks = [0,10,20]
yticklabels = [0,10,20]
xticks = [0,24,4*12,6*12,8*12,10*12,12*12, 14*12, 16*12,18*12, 20*12,22*12,24*12,25*12]
xticklabels = ['1992','1994', '1996','1998', '2000', '2002', '2004', '2006', '2008',  '2010','2012', '2014', '2016', '2017']
plt.xticks(xticks, xticklabels, size = 20)
plt.yticks(yticks, yticklabels, size = 20)
plt.xlabel("Year", size=30)
plt.ylabel("AMOC (Sv)", size=30)
plt.title('AMOC at 10 degree from 1992 to 2017', size=40)
plt.gca().xaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())

plt.show()
```
